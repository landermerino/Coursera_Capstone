## 1. Introduction & Business Problem

**Problem Background**  
Basque Country is an autonomy located at the north of Spain, it is divided in three boroughts Bizkaia, Guipuzkoa and Vitoria. The population in this autonomy is diverse and it provides a lot of business opportunities. The city is a major center for turism, retailing, transportation, culture and fashion in Spain.
The economy of Basque Country has exponentialy grow up, becoming a highly competitive place to start a business. For this reason, any new businnes venture needs to be analysed with the avaliable data. This would give a good insight of the necessities of the business and reduct the risks.

**Problem Description**  
A restaurant is a business which serves food and drinks in return for money. Basque Country is famous for its excellent local cusine. But the increasement of the population and turism has bring an opportunitie to open new kind of restaurants. For this activity I will make an analysis for the best city to open a new restaurant, looking the different nationalities of the population, the density and the economic power of the citizents.
